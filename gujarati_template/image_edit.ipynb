```python
def create_gujarati_text(text, font_path, font_size, color):
    dummy = Image.new("RGBA", (1, 1))
    draw = ImageDraw.Draw(dummy)
    font = ImageFont.truetype(font_path, font_size)
    bbox = draw.textbbox((0, 0), text, font=font)

    w = bbox[2] - bbox[0]
    h = bbox[3] - bbox[1]

    img = Image.new("RGBA", (w, h), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)
    draw.text((0, 0), text, font=font, fill=color)
    return img


base = Image.open("input_images/haldi_.png").convert("RGBA")

text_img = create_gujarati_text(
    text="હલ્દી સમારોહ",
    font_path="Fonts/NotoSansGujarati-VariableFont_wdth,wght.ttf",
    font_size=80,
    color=(255, 204, 0, 255)
)

base.paste(text_img, (120, 300), text_img)
base.save("merged_final.png")
```

```python
from PIL import Image, ImageDraw, ImageFont

# ---------- Helper function ----------
def create_gujarati_text(text, font_path, font_size, color):
    dummy = Image.new("RGBA", (1, 1))
    draw = ImageDraw.Draw(dummy)
    font = ImageFont.truetype(font_path, font_size)

    bbox = draw.textbbox((0, 0), text, font=font)
    w = bbox[2] - bbox[0]
    h = bbox[3] - bbox[1]

    img = Image.new("RGBA", (w, h), (0, 0, 0, 0))
    draw = ImageDraw.Draw(img)
    draw.text((0, 0), text, font=font, fill=color)
    return img


# ---------- Load base image ----------
base = Image.open("input_images/haldi_blank.png").convert("RGBA")

FONT_PATH = "Fonts/NotoSansGujarati-VariableFont_wdth,wght.ttf"
GOLD_COLOR = (255, 215, 140, 255)   

# ---------- Title ----------
title_img = create_gujarati_text(
    text="હલ્દી સમારોહ",
    font_path=FONT_PATH,
    font_size=90,
    color=GOLD_COLOR
)

# Center title horizontally
title_x = (base.width - title_img.width) // 2
title_y = 120
base.paste(title_img, (title_x, title_y), title_img)


# ---------- Details ----------
details = [
    "તારીખ: ૨૮ સપ્ટેમ્બર ૨૦૨૫",
    "સમય: સવારે ૧૧:૦૦ વાગ્યે",
    "સ્થળ: રુક્મણી પેલેસ"
]

start_y = base.height - 280
line_gap = 70

for i, line in enumerate(details):
    line_img = create_gujarati_text(
        text=line,
        font_path=FONT_PATH,
        font_size=56,
        color=GOLD_COLOR
    )

    x = (base.width - line_img.width) // 2
    y = start_y + i * line_gap
    base.paste(line_img, (x, y), line_img)


# ---------- Save ----------
base.save("merged_final_gujarati.png")
```
